## ‚úçÔ∏è Exercici 1: Ordenar la llista de vols per pa√≠s

L'Ajuntament de Barcelona est√† estudiant el desplegament d'un servei d'informaci√≥ per a turistes internacionals. Per a la prova pilot nom√©s pot contractar un equip d'informadors que tingui fluidesa en quatre idiomes estrangers diferents, aix√≠ que vol assegurar-se que es tracta dels idiomes estrangers que parlen el m√†xim nombre de turistes que arriben a l'aeroport de Barcelona.

Defineix una funci√≥ que implementi un algoritme d'ordenament r√†pid o quicksort (amb les seves funcions auxiliars necess√†ries "particio" aix√≠ com tamb√© la funci√≥ quicksort recursiva) i que donada una llista desordenada de tuples amb un n√∫mero i la seva etiqueta (n√∫mero de vols i pa√≠s) et retorni una llista amb les tuples ordenades pel n√∫mero.

Hi ha moltes variants de l'algoritme d'ordenament r√†pid o quicksort, per√≤ t√© aquests passos b√†sics:
1) Triar un element de la llista que sigui el pivot. Hi ha diferents estrat√®gies per a fer aix√≤.
2) Posar tots els elements m√©s grans que el pivot a la seva dreta i els m√©s petits a l'esquerra, sense preocupar-se de l'ordre d'aquests elements. Aix√≤ deixar√† el pivot a la seva posici√≥ ordenada dins la llista. Aquesta part de l'algoritme es diu "partici√≥" i tamb√© hi ha moltes maneres de fer-ho.
3) Repetim el proc√©s iterativament tant pel tros de llista que queda a l'esquerra del pivot com pel tros de llista que queda a la dreta.

Pots fer un cop d'ull a aquest v√≠deo per repassar un algoritme d'ordenament r√†pid o quicksort:

```python
import random
# Comprova que la teva implementaci√≥ de l'algoritme d'ordenaci√≥ r√†pida o
# quicksort ordena b√© aquesta llista de tuples amb el n√∫mero de vols i pa√Øsos:
llista_vols = [(41, "Nova Zelanda"), (93, "Col√≤mbia"), (34, "Brasil"), 
               (0, "Narnia"), (20, "Madagascar"), (12, "Ecuador")]
```

```python
def particio_tuples(llista:list[tuple[int, any]], primer: int, darrer: int) -> int:
    """Posa el pivot en la seva posici√≥ ordenada dins un tros de llista (slice).

    Aquesta funci√≥ considera els elements de la llista compresos entre el primer
    i el darrer √≠ndexs indicats, 
    defineix el primer com el pivot, 
    i el situa en la posici√≥ que tindria dins d'aquest tros de llista si els elements
    estiguessin ordenats. Tots els elements amb valors m√©s petits obtenen
    posicions amb √≠ndexs m√©s baixos que el pivot i els que tenen valors m√©s grans
    obtenen √≠ndexs m√©s alts, tot i que aquests elements m√©s petits i m√©s grans no
    tenen perqu√® estar ordenats entre s√≠. Els elements de la llista amb √≠ndexs
    m√©s baixos que el pivot no es modifiquen. Aqu√≠ s'implementa l'algoritme de
    partici√≥ de Hoare, que √©s el m√©s r√†pid.

    En l'algoritme d'ordenaci√≥ r√†pida o quicksort, aquest pas de partici√≥ es pot
    repetir diverses vegades i obtenir una llista ordenada.

    A m√©s, en aquesta implementaci√≥ els elements de la llista no s√≥n simples
    nombres, sin√≥ que s√≥n tuples que s'ordenen tenint en compte nom√©s el primer
    element, ja que el segon element es considera un identificador o etiqueta.
    Aquesta funci√≥ no genera una llista nova sin√≥ que modifica la d'entrada i
    retorna l'√≠ndex que tindria el pivot en la llista si el tros de llista
    especificat estigu√©s ordenat.

    Par√†metres:
    -----------
    llista (llista de tuples): Llista de tuples potencialment desordenades.
    primer (int): Primer √≠ndex del tros de la llista que es t√© en compte (slice),
        i tamb√© √≠ndex del pivot o tupla de refer√®ncia que es situar√† en la seva
        posici√≥ ordenada dins d'aquest tros de llista.
    darrer (int): Darrer √≠ndex del tros de la llista que es t√© en compte.

    Retorna:
    --------
    √çndex del pivot o tupla de refer√®ncia en la llista si s'orden√©s aquest tros.

    Once this function ends, the list is modified as:
    (llista[0][0] to llista[index][0]) <= llista[0][0] <= (llista[index + 1][0] to llista[-1][0]])
    """

    pivot: int = llista[primer][0]
    left = primer - 1
    right = darrer + 1

    while True:
        left += 1
        while llista[left][0] < pivot:
            left += 1
        
        right -= 1

        while llista[right][0] > pivot:
            right -= 1
        
        if left >= right: return right
        llista[left], llista[right] = llista[right], llista[left]
```

```python
def quicksort_tuples_r(
    llista: list[tuple[int, any]], primer: int, darrer: int
) -> None:
    """Partici√≥ recursiva de les llistes a cada banda del pivot (quicksort).

    Aquesta funci√≥ implementa un algoritme recursiu de quicksort triant sempre
    com a pivot el primer element de la llista. Un cop ha trobat la posici√≥ del
    pivot, torna a executar aquesta mateixa funci√≥ pel tros de la llista amb
    valors m√©s petits que pivot i pel tros de la llista amb valors m√©s grans al
    pivot, fins que el pivot √©s l'√∫nic element.
    
    Par√†metres:
    -----------
    llista (llista de tuples): Llista de tuples potencialment desordenades.
    primer (int): Primer √≠ndex del tros de la llista que es t√© en compte (slice),
        i tamb√© √≠ndex del pivot o tupla de refer√®ncia que es situar√† en la seva
        posici√≥ ordenada dins d'aquest tros de llista.
    darrer (int): Darrer √≠ndex del tros de la llista que es t√© en compte.

    Retorna:
    --------
    None
    """
    if (primer >= darrer):
        return llista[primer]
    
    pivot = particio_tuples(llista, primer, darrer)
    quicksort_tuples_r(primer, pivot)
    quicksort_tuples_r(pivot + 1, darrer)
```

```python
def quicksort_tuples(llista: list[tuple[int, any]]) -> list[tuple[int, any]]:
    """Ordena una llista de tuples pel primer element de cada tupla.
    
    Aquesta funci√≥ ordena en ordre ascendent una llista de tuples pel n√∫mero que
    apareix com a primer element de la tupla, tot usant l'algoritme d'ordenaci√≥
    de Quicksort. El segon element de la tupla serveix per identificar l'element
    i pot tenir qualsevol altre valor, per√≤ no es t√© en compte per a la
    ordenaci√≥ de les tuples.
    
    Par√†metres:
    -----------
        llista_desordenada: Llista de tuples desordenades, on el primer element 
        de cada tupla √©s un n√∫mero i el segon element es una etiqueta qualevol.

    Retorna:
    --------
        Llista de tuples ordenades en ordre ascendent pel n√∫mero que apareix com
        a primer element de cada tupla, i on el segon element √©s una etiqueta o
        identificador qualsevol d'aquell element.
    """
    
    not_sorted: list[tuple[int]] = [(0, len(llista) - 1)]

    while (not_sorted):
        last_interval: tuple[int] = not_sorted[-1]
        not_sorted.pop()

        left, right = last_interval
        if left >= right:
            continue

        pivot: int = particio_tuples(llista, left, right)
        not_sorted.append((left, pivot))
        not_sorted.append((pivot + 1, right))
```

```python
# L'algoritme quicksort no genera una llista nova sin√≥ que reordena l'existent
quicksort_tuples(llista_vols)

# Comprovem que l'algoritme ha ordenat la llista de tuples
assert(llista_vols == [(0, "Narnia"), (12, "Ecuador"), (20, "Madagascar"),
                       (34, "Brasil"), (41, "Nova Zelanda"), (93, "Col√≤mbia")])
```

## ‚úçÔ∏è Exercici 2: D'on provenen la major part de turistes?

A l'arxiu "2024_TransitAeri_Pais.csv" hi pots trobar una llista dels vols que van arribar cada dia a l'aeroport de Josep Tarradellas Barcelona-El Prat durant l'any 2024 provinents dels diferents pa√Øsos del m√≥n.

Fes una funci√≥ que llegeixi l'arxiu de dades, sumi el n√∫mero total de vols provinents de cada pa√≠s i que, utilitzant la implementaci√≥ de l'ordenaci√≥ amb Quicksort que has codificat a l'exercici anterior, et doni una llista dels pa√Øsos dels quals han arribat m√©s vols a l'aeroport. Assegura't que la teva funci√≥ exclou Espanya d'aquesta llista, ja que a l'Ajuntament nom√©s li interessen els turistes internacionals.

```python
from typing import Any
import random
def paisos_mes_vols() -> list[tuple[int, Any]]:
    """
    D√≥na una llista ordenada amb els pa√Øsos dels que aterren m√©s vols.

    Aquesta funci√≥ llegeix un arxiu de dades amb la llista de vols diaris
    provinents de cada pa√≠s, i d√≥na una llista ordenada en ordre descendent dels
    pa√Øsos estrangers dels que aterren m√©s vols anualment (excloent Espanya).

    Retorna:
    --------
    Llista de tuples ordenades en ordre descendent pel n√∫mero de vols totals que
    aterren provinents de cada pa√≠s, on el primer element de cada tupla √©s el
    n√∫mero total de vols i el segon element de la tupla √©s el nom del pa√≠s.

    date, code, name, number
    0. read file, make code:country
    1. read file
        make count_tourists
    2. convert the dict to a list
    3. sort the list
    """

    set_country = get_countries_from_first_day()
    d_country_people: dict[str, int] = filling_country_people(set_country)

    l_country_people: list[tuple[int, str]] = [(people, country) for country, people in d_country_people.items()]
    decreasing_sort(l_country_people, 0, len(l_country_people) - 1)

    return l_country_people


def get_countries_from_first_day() -> set[str]:
    set_country: set[str] = set()
    with open("2024_TransitAeri_Pais.csv", "r") as file:
        file.readline()  # skip the header

        prev_message: list[Any] = file.readline().strip().split(",")
        while True:
            set_country.add(prev_message[2])

            curr_message: list[Any] = file.readline().strip().split(",")
            if prev_message[0] != curr_message[0]:
                break

            prev_message = curr_message
        
        return set_country
    

def filling_country_people(set_country: set[str]) -> dict[str, int]:
    country_people: dict[str, int] = {country:0 for country in set_country}

    with open("2024_TransitAeri_Pais.csv", "r") as file:
        file.readline()  # skip the header

        for line in file:
            message: list[Any] = line.strip().split(",")
            country: str = message[2]
            people: int = int(message[3])

            country_people[country] += people

    del country_people["Espanya"]
    return country_people


def decreasing_sort(l: list[tuple[int, str]], left: int, right: int) -> None:
    if left >= right:
        return
    
    pivot = partition(l, left, right)
    decreasing_sort(l, left, pivot)
    decreasing_sort(l, pivot + 1, right)


def partition(l: list[tuple[int, str]], left: int, right: int) -> int:
    pivot = l[random.randint(left, right)][0]

    i = left - 1
    j = right + 1
    while True:
        i += 1
        while l[i][0] > pivot:
            i += 1
        j -= 1
        while l[j][0] < pivot:
            j -= 1
        
        if i >= j: 
            return j
        l[i], l[j] = l[j], l[i]
```

## ‚úçÔ∏è Exercici 3: Complexitat dels algoritmes d'ordenaci√≥

Observa el funcionament dels algoritmes de Quicksort i de Mergesort. Quina complexitat t√© cadascun? Quin √©s m√©s eficient?

La complexidad del quicksort en situaci√≥n general es O(n * logn), pero ser√° O(n^2) en la situaci√≥n peor  
La del mergesort es O(n * logn) en todos los casos, pero es un poco m√°s lento que el O(n * logn) del quicksort en situaci√≥n general  
  
Generalmente, el quicksort es m√°s eficiente, pero un poco inestable.
