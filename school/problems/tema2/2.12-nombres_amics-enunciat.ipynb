### 2.12 Nombres amics

Diem que dos nombres són amics si la suma dels divisors d'un d'ells és equivalent a l'altre i viceversa. Per exemple:
- Els divisors de $220$ són $1,2,4,5,10,11,20,22,44,55,110$ que sumen $284$.
- Els divisors de $284$ són $1,2,4,71, 142$, que sumen $220$.

Per tant, $220$ i $284$ són amics. Escriu una funció que determini si una parella de nombres són amics. Fes servir aquesta plantilla per escriure el teu codi.

```python
import math
def sum_divisors(num: int) -> int:
    """
    This function returns the sum of the divisors of num

    Paremeters
    ----------
    num: int
        The number whose sum of divisors is to be calculated

    Returns
    ---------
        The sum of divisors of num
    """
    max_divisor_possible:int = int(math.sqrt(num))
    list_divisor: list[int] = [1];

    for i in range(2, max_divisor_possible + 1):  # Complexity O(sqrt(n))
        if num % i == 0:
            list_divisor.append(i)
            list_divisor.append(int(num / i))

    if list_divisor[-1] == math.sqrt(num): list_divisor.pop()

    # Every divisor of num is now in list_divisor
    return sum(list_divisor)
```

```python
def amics(num1:int, num2:int) -> bool:
    """
    Aquesta funció comprova si dos nombres són amics.
    
    Parameters
    ----------
    num1: int
        El primer número
    num2: int
        El segon números
    
    Returns
    -------
    b: bool
        Si la parella de nombres d'entrada són amics    
    """
    return (sum_divisors(num1) == num2) and (num1 == sum_divisors(num2))
```

```python
assert amics(220, 284) == True
assert amics(200, 230) == False
assert amics(9363584, 9437056) == True
```
