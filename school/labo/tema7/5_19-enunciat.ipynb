### 锔 Problema 5.19 Karatsuba

Escriu una funci贸 recursiva que calculi la multiplicaci贸 pel m猫tode de Karatsuba en base 10 de dos nombres.

Karatsuba 茅s un algorisme que optimitza la multiplicaci贸 aritm猫tica, reduint les opreacions a fer:

Sigui $A = A_1 10^{n/2} + A_0$ amb $A_0 < 10^{n/2}$

Sigui $A = B = B_1 10^{n/2} + B_0$ amb $B_0 < 10^{n/2}$

Llavors $C = A * B = (A_1 10^{n/2} + A_0)(B_1 10^{n/2} + B_0)$ 
es pot calcular com: $A_1 B_1 10^{n} + (A_1 B_0 + A_0 B_1) 10^{n/2} + A_0 B_0$

Exemple: $2925 \times 6872 = (29 \times 10^2 + 25) \times (68 \times 10^2 + 72) = 29   \times 68 \times 10^4 + (29 \times 72 + 25 \times 68) \times 10^2 + 25 \times 72$


Si ens fixem el terme del mig el podem posar en funci贸 dels altres dos: 
- Sabem que $(a+b)(c+d) = ac + ad + bc + bd$
- Si ja tenim ac i bd calculats, podem dir que $(ad + bc) = (a+b)(c+d) - ac - bd$


El descriurem pas a pas:

1. Descomposar el primer sumand, $x$,  en dues parts  $x1$ i $x2$, cadascuna amb la meitat de d铆gits
2. Descomposar el segon sumand, $y$, en dues parts $y1$ i $y2$, cadascuna amb la meitat de d铆gits
3. Calcular $x1*y1$,  $x1y1$, crida recursiva
4. Calcular $x2*y2$, $x2y2$, crida recursiva
5. Calcular el terme del mig $x1*y2 + x2*y1$ com a $(x1 + x2)*(y1+y2)-x1y1-x2y2$, aquesta part $(x1 + x2)*(y1+y2)$ la fem com a crida recursiva
6. Multiplicar cada terme per la pot猫ncia de 10 corresponent i sumar i ja tenim la multiplicaci贸:
- $x1y1 * 10^m$ + $(x1*y2 + x2*y1) * 10^{m/2} $ + $x2y2$

```python

# Ho veiem amb uns n煤meros concrets
x = 2925
y = 6875
x1 = 29
x2 = 25
y1 = 68
y2 = 75
x1y1 = 1972   # x1 * y1, 29*68
x2y2 = 1875   # x2 * y2, 25*75
x1mesx2 = 54  # x1 + x2, 29 + 25
y1mesy2 = 143  # y1 + y2, 68 + 75
termemig = 3875  # x1mesx2 * y1mesy2 - x1y1 - x2y2, 54 * 143 - 1972 - 1875
xy = 20109375 # x1y1 * 10000 + termemig * 100 + x2y2
```

```python
def karatsuba(x,y):
    """
    Aquesta funci贸 implementa l'algorisme de multiplicaci贸 de Karatsuba de manera recursiva.

    Parameters
    ----------
    x: int de 4 d铆gits
    y: int de 4 d铆gits

    Returns
    -------
    prod: int

    Process
    ---
    x = x1 + x2
    y = y1 + y2
    x * y = (
        (x1 * y1)  * 10^n +
        ((x1 + x2) * (y1 + y2) - x1 * y1 - x2 * y2) * 10^(n/2) +
        (x2 * y2)
    )

    x1mesx2 * y1mesy2 - x1y1 - x2y2
    """
    if (x < 100 or y < 100): 
        return x * y

    m: int = len(str(x))  #________________________________________________O(n)
    x1, y1 = x // pow(10, m / 2), y // pow(10, m / 2)  #___________________O(1)
    x2, y2 = x % pow(10, m / 2), y % pow(10, m / 2)

    first: int = karatsuba(x1, y1)
    last: int = karatsuba(x2, y2)
    middle_term: int = karatsuba((x1 + x2), (y1 + y2)) - first - last

    prod: int = first * pow(10, m) + middle_term * pow(10, m / 2) + last  # O(1)

    return prod

# La dificuldad sin contar la llamada recursiva es O(n) == O(n^1)
```
