$$ \{(L[i], L[j]) | i < j, L[i] + L[j] = s, i, j \in \{0, \ldots, n-1\}\} $$

Per exemple:
\
Si la funci√≥ rep una llista ordenada que t√© elements **[2, 5, 6, 7, 8]** i un valor **13**, la funci√≥ hauria de retornar la llista **[(5,8), (6,7)]**, on cada tupla representa una parella de nombres que sumen **13**. Cada tupla segueix que el primer nombre √©s el m√©s petit de la parella. Si tenim els nombres 7 i 6, necess√†riament la tupla ha de ser **(6,7)**.
\
