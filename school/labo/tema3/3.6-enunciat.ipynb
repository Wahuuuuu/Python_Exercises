## 锔 Problema 3.16: Aritmetica modular. NIF

Escriu una funci贸 que, donat un n煤mero en el format adient d'un NIF, validi si la lletra es correspon realment al n煤mero del DNI. Has de fer servir alguna col路lecci贸 de Python que sigui adequada. 

```python
def validar_NIF(cadenaNIF:str) -> bool:
    """
    Aquesta funci贸 valida si la lletra correspon al DNI
    
    Parameters
    ----------
    cadenaNIF: str
        NIF
        
    Returns
    -------
    esCorrecte: bool
        Retorna si el NIF 茅s correcte o no.
    """
    list_char: list[str] = [i for i in "TRWAGMYFPDXBNJZSQVHLCKE"]         # O(1)
    dic_nif: dic[int, str] = {i : j for i, j in enumerate(list_char, 0)}  # O(1)

    n_nif: int = int(cadenaNIF[:-1])                                      # O(1), porque la len(cadenaNIF) no varia
    return (dic_nif[n_nif % 23] == cadenaNIF[-1])                         # O(1)
```

```python
#   per provar...

assert validar_NIF('56789123F') == True
assert validar_NIF('56789123H') == False
```

## 锔 Problema 3.16: Aritmetica modular. Fulla de clcul

Escriu una funci贸 per convertir un nombre en el nom de la columna del full de clcul corresponent.

Ex. 1 => A, 11 => K

```python
from typing import Any

def conversio_fulla_calcul(num:int) -> str:
    """
    Aquesta funci贸 tradueix el valor num en el nom corresponent que tindriem en un full de clcul.
    
    Parameters
    ----------
    num: int
    
    Returns
    columna: str
    """                       
    if (num == 0): 
        raise ValueError("El numero no puede ser zero.")

    char = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    quocient: int = num
    resta: int
    digit: list = []
    while (quocient > 0):     # Complexitat O(log n)
        quocient -= 1
        quocient, resta = divmod(quocient, 26)
        digit.append(char[resta])
    digit.reverse()           # Complexitat O(1)
    
    return "".join(digit)     # Complexitat O(1)
```

```python
#   per provar...

conversio_fulla_calcul(2)
```
