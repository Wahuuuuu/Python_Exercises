## 锔 Problema 3.2: Escriu una funci贸 que simplifiqui una fracci贸 a la seva expressi贸 irreductible (fes servir gcd)

````python
from typing import Any

def gcd(a,b):
    while a:
        a, b = b%a, a
    return b

def reduir_fraccio(numerador:int, denominador:int) -> Any:
    """
    Funci贸 que retorna l'expressio irreductible d'una fracci贸
    
    Parameters
    ----------
    numerador: int
    denominador: int
    
    Returns
    -------
    numReduit: int
    denReduit: int
    """
    mcd: int = gcd(numerador, denominador)
    return (numerador / mcd, denominador / mcd)
    
````

````python
#   per provar...

assert reduir_fraccio(12, 8) == (3,2)
````
