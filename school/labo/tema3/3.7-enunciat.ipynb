```python
def calculadora():
    """
    Aquesta funci贸 representa una calculadora.
    Llegueix l'entrada per linia de comandes.
    
    Parameters
    ----------
    None
    
    Returns
    -------
    None
    """
    operation: str = input("Entra operaci贸: ") # O(n)
    have_input: bool = operation
    while have_input:
        # found the operator
        i:int = 0
        while operation[i] not in "+-*":       # O(n)
            i += 1

        operator: str = operation[i]
        if operator == "+": print(int(operation[:i]) + int(operation[i+1:]))
        elif operator == "-": print(int(operation[:i]) - int(operation[i+1:]))
        else: print(int(operation[:i]) * int(operation[i+1:]))
                
        operation = input("Entra operaci贸: ")
        have_input = operation
```

## 锔 Problema 3.7: Nombre m茅s petit divisible

El nombre 2520 茅s el nombre m茅s petit que es pot dividir de manera exacta (sense decimals) per cada un dels nombres enters entre 1 i 10. Escriu un algorisme que calculi el nombre m茅s petit divisible per tots els nombres menors a un nombre $n$ donat.

```python
def mes_petit_divisible_per(n:int) -> int:
    """
    Aquesta funci贸 calcula el nombre m茅s petit divisible per tots els nombres menors a numero n
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    :int
    """

    l:list[bool] = [True] * (n+1)  # O(n)
    l[0] = False
    l[1] = False

    lcm: int = 1
    # Buscar numero primo i
    for i in range(len(l)):        # O(n)
        if l[i] == True:

            max_potencial: int = 1;
            while max_potencial * i <= n:  # O(log n)
                max_potencial *= i
            lcm *= max_potencial

            # Borrar m煤ltiplos de i de la lista
            for j in range(i, n+1, i): # O(log n)
                l[j] = False

    return lcm
```

```python
###   Per provar...

assert mes_petit_divisible_per(2) == 2
assert mes_petit_divisible_per(4) == 12
assert mes_petit_divisible_per(9) == 2520
assert mes_petit_divisible_per(15) == 360360
```
